```python
# Supondo que 'base' é o nome do seu DataFrame
coluna_silte = base['SILTEAR']

# Verificando valores faltantes na coluna 'silte'
valores_faltantes = coluna_silte.isnull().sum()

print(f"Valores faltantes na coluna 'silte': {valores_faltantes}")
```

```python
base.iloc[:, 5] = base.iloc[:, 5].str.replace(',', '.').astype(float)
```

```python
x = base.iloc[:, 1].values #Areia
y = base.iloc[:, 3].values # Argila
correlacao = np.corrcoef(x, y)
correlacao
#negativo positivo 
```

```python
x = x.reshape(-1, 1)
```

```python
#criação de modelo
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
#interceptaçãp
modelo.intercept_
```

```python
#inclinação
modelo.coef_
```

```python
#valor de r2
#89% da variável dependente pode ser explicada pela variável explanatória do modelo 
modelo.score(x,y)
```

```python
previsoes = modelo.predict(x)
previsoes
```

```python
#criação de outro modelo uzando a biblioteca statsmodel
modelo_ajustado = sn.ols(formula = 'ARGILA ~ AREIA', data = base)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
plt.scatter( x,y)
plt.plot(x, previsoes, color = 'red')
plt.title( 'areia x argila')
plt.xlabel('areia')
plt.ylabel(' argila' )
```

```python
# se um perfil tiver 200 de areia qual seria sua quantidade de argila
modelo.predict([[200]])
```
